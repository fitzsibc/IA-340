````python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
````

````python
import matplotlib.pyplot as plt

# Assuming the dataframe has a 'Year' column and an 'Income' column
df_va = df.sort_values(by='Year')

# Calculate the income growth rate
df_va['Income_Growth_Rate'] = df_va['Income'].pct_change() * 100

# Create a line chart to display income growth rate over the years
plt.figure(figsize=(10,6))
plt.plot(df_va['Year'], df_va['Income_Growth_Rate'], marker='o')
plt.title('Income Growth Rate in VA Over the Years')
plt.xlabel('Year')
plt.ylabel('Income Growth Rate (%)')
plt.grid(True)
plt.show()
````

````python
# Assuming 'income' column in the dataframe represents the income and 'year' column represents the year

df['year'] = pd.to_datetime(df['year'])
df.set_index('year', inplace=True)

# Calculate yearly income
yearly_income = df['income'].resample('A').sum()

# Calculate the growth rate
income_growth_rate = yearly_income.pct_change()

# Convert it to percentage
income_growth_rate = income_growth_rate * 100

income_growth_rate
````
